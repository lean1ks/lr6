```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Администратор\\Desktop\\ЛР6\\Данные.csv', sep='\t')
df
```

# Заменяем пропуски

```python
df['balcony'] = df['balcony'].fillna(0)
df
```

```python
df['is_apartment'] = df['is_apartment'].fillna(0)
df
```

```python
#df['ceiling_height'] = df['ceiling_height'].fillna( )
#df

ka=df['kitchen_area'].median()
df['kitchen_area'] = df['kitchen_area'].fillna(ka)
df
```

# Cчитаем и добавляем в таблицу:

# цену кв метра

```python
df['price_metr'] = df['last_price']/df['total_area']
df['price_metr'] = df['price_metr'].astype(int)
df
```

```python
df['first_day_exposition']=pd.to_datetime(df['first_day_exposition'])
df
```

```python
df['price_metrprice_metr'] = df['price_metr'].astype(int)
df
```

# •день недели, месяц и год публикации объявления

```python
df['day'] = pd.DatetimeIndex(df['first_day_exposition']).day
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year
df
```

# •	этаж квартиры; варианты — первый, последний, другой

```python
fl_t = df['floors_total'].median()
df['floors_total'] = df['floors_total'].fillna(fl_t)
df['floors_total'] = df['floors_total'].astype(int)
def cat_floor(column):
    if column[8] == '1':
        return 'first_floor'
    elif column[8] == column[6]:
        return "last"
    else:
        return 'another'
df['kakoy_etazh'] = df.apply(cat_floor, axis =1)
df
```

# •	соотношение жилой и общей площади, а также отношение площади кухни к общей

```python
k_a =df['kitchen_area'].median()
df['kitchen_area'] = df['kitchen_area'].fillna(k_a)
l_a=df['living_area'].median()
df['living_area'] = df['living_area'].fillna(l_a)
df['living_total'] = df['living_area']/df['total_area']
df['kitchen_total'] = df['kitchen_area']/df['total_area']
df
```

```python
df.describe()
```

```python
df['rooms']=df['rooms'].replace(0,1)
```

# Изучаем след параметры:

# площадь, цена, число комнат, высота потолков. Строим гистограммы для каждого параметра

```python
plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
plt.hist(df['total_area'],bins=60,color='skyblue',edgecolor='black')
plt.title('Гистограмма площади')

plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
plt.hist(df['last_price'],bins=100,color='green',edgecolor='black')
plt.title('Гистограмма price')

plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
plt.hist(df['rooms'],bins=50,color='orange',edgecolor='black')
plt.title('Гистограмма rooms')

plt.figure(figsize=(12,12))
plt.subplot(2,1,2)
plt.hist(df['ceiling_height'],bins=50,color='red',edgecolor='black')
plt.title('Гистограмма высота потолков')



plt.tight_layout()
plt.show
```

# Гистограмма время продажи квартиры

```python
d_e = df['days_exposition'].median()
df['days_exposition'] = df['days_exposition'].fillna(d_e)
plt.bar(df['first_day_exposition'],df['days_exposition'], align='center', alpha=0.5, width=10,color = 'gray')
plt.xticks(rotation = 45)
plt.ylabel('продолжительность')
plt.xlabel('дата размещения')
plt.title('отношение продолжительности и даты')
plt.grid()
plt.show()
```
